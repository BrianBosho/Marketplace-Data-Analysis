```python
# Define a function that can extract the desire key value pairs that are attributes of the product
def extract_attrs(x):
    ls_x=[]
    for item in x:
        try:
            ls_x.append({item['name']:item['value']})
        except:
            ls_x.append(x)        
    return ls_x
```

```python
# define a function that extracts the attrubutes (keys)
def extract_keys(x):
    attrs_set=[]
    for item in x:
        try:
            attrs_set.append(item['name'])
        except:
             attrs_set.append(x)      
    return  attrs_set
```

```python
# get the attribute key value pairs as a list
ls_attrs = df_ads['attrs'].apply(extract_attrs)
# convert the key value pairs from a list of lists to a list of dicts
attrs_dicts=[]
for item in ls_attrs:
    ls_attrs5={}
    for d in item:
        ls_attrs5.update(d)
    attrs_dicts.append(ls_attrs5)
```

```python
# get the attribute key value pairs as a list
keys_attrs=df_ads['attrs'].apply(extract_keys)
ls_keys=[]
for item in keys_attrs:
    ls_keys.extend(item) 
    
# convert the list to a set so that it has unique values to be used as df column names
set_keys = set(ls_keys)
# convert the set now to a list of unique items
ls_keys= list(set_keys)
```

```python
df_attrs=pd.DataFrame.from_dict(attrs_dicts)
```
