```python
dfs = [df_attr_basic, df_price_data, df_seller_data, df_adverts_basic]
final_df = reduce(lambda  left,right: pd.merge(left,right,left_index=True, right_index=True, how='outer'), dfs)
final_df.head()
```

```python
min_est = final_df.price_estimate[0].split("-")[0].split(" ")[1]
min_est
```

```python
max_est = final_df.price_estimate.values[0].split("-")[1].split(" ")[1]
max_est
```

```python
final_df['min_est']=final_df['price_estimate'].apply(lambda x:x if type(x)==float else x.split("-")[0].split(" ")[1])
```

```python
def get_max(x):
    try: 
        return x if type(x)==float else x.split("-")[1].split(" ")[1]
    except:
        return x
    
```

```python
final_df['max_est']=final_df['price_estimate'].apply(lambda x:get_max(x))
```

```python
final_df['min_est']=final_df['min_est'].astype(float)
```

```python
final_df['min_est'].max()
```

```python
final_df['min_est']=final_df['min_est'].apply(lambda x:x/1000 if x>45 else x)
final_df['min_est']=final_df['min_est'].apply(lambda x:float(x)*1000000)
```

```python
def to_float(x):
    try:
        return float(x)*1000000
    except:
        float(x.split(' ')[1])*1000000
```

```python
final_df['max_est']=final_df['max_est'].apply(lambda x:to_float(x))
final_df['max_est']=final_df['max_est'].apply(lambda x:x/1000 if x>100000000 else x)
```

```python
final_df['avg_est']= (final_df.min_est+final_df.max_est)/2
```
