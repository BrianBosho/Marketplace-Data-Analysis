```python
#basic bar chart
x=df[:20]['age']
y=df[:20]['listing_duration']

#use plt.bar to plot a bar plot
plt.bar(x,y)
```

```python
# Create a df from value counts so that we can plot
df_makes= df.Make.value_counts().rename_axis('Make').to_frame('counts').reset_index()
df_makes
```

```python
#regular vertical barchart
x=df_makes[:10]['Make']
y=df_makes[:10]['counts']

plt.xlabel('Makes')
plt.ylabel('Count')
plt.title('Popular makes')

plt.xticks(rotation=90)
plt.bar(x, y)
plt.show()
```

```python
#using barh to make a horizontal bar chart
df_models= df.Model.value_counts().rename_axis('Model').to_frame('Counts').reset_index()
df_models


x=df_models[:30]['Model']
y=df_models[:30]['Counts']

plt.xlabel('Model')
plt.ylabel('Count')
plt.title('Popular Models')

plt.xticks(rotation=90)
plt.bar(x, y)

plt.figure(figsize=(18, 16))
plt.show()
```

# Ploting pie charts

```python
#simple piechart
data =df_makes[:5]['counts']
labels=df_makes[:5]['Make']
plt.pie(data, labels=labels)
plt.show()
```

```python
#Pie chart of the top 10 makes using plt.pie
data =df_makes[:5]['counts']
labels=df_makes[:5]['Make']

#using explode to put emphasis on a specific item
explode=[0,0,0.2,0,0]

# plt.pie(data,labels=labels) # two lists used to make the pie chart

# Additional properties (wedges)
# Use shadow to give a 3D effect
#use starting angle to start and set position
#show percentage using: autopct='%1.1f%%'
plt.pie(data,labels=labels, explode=explode, shadow=True, startangle=0, autopct='%1.1f%%', wedgeprops={'edgecolor':'white'})

plt.title('Top 5 makes')
plt.show()
```

# Histograms

```python
df['selling_price'].min()
```

```python
# create a filter for minimum and maximum price to remove outliers
sp_filt= (df['selling_price']<5100000) & (df['selling_price']>=100000)
# get values to plot histogram
values = df.loc[sp_filt]['selling_price']
# use the bins argument to specify the number of bins
# use the edcolor to get the delniating color
# use log when there are extreme values
# get the median: And draw a vertical line there
median_price = values.median()
mean_price = values.mean()
plt.axvline(median_price, color='black', label= 'median price')
plt.axvline(mean_price, color='red', label= 'mean price')
#title
plt.title('Price distribution')
# Show legends
plt.legend()
plt.hist(values, bins=40, edgecolor='white', log=True)
plt.show()
```

# Scatter plots

```python
# Basic plot
# create a filter to contrain the plot to a specific model
model_filt = (df['Model']== 'Harrier') & (df['Mileage'] <400000) & (df['selling_price'] <2500000)

#data to plot
dfp = df.loc[model_filt]
y=dfp[:50]['selling_price']
x=dfp[:50]['Mileage']

#doing the scatter plot
plt.scatter(x,y)

#show the plot
plt.show()
```

```python
# scatter plot with customizations

# doing the scatter plot
# set the size of the dots using the argument s
# specify the color with the c 
# you can specify the marker if you d0 not want to use the dots
# add edgecolor
plt.xticks(rotation=90)
plt.scatter(x,y, s=100, c='green', edgecolor='black', alpha = .75)

#show the plot
plt.show()
```

```python
# Additional modifications

#use age of the car to give color to the dot
age =dfp[:50]['age']

plt.xticks(rotation=90)
plt.scatter(x,y, s=100, c=age, cmap = 'Greens', edgecolor='black', alpha = .75)

#show the plot
plt.show()
```

```python
# using color bar

#use age of the car to give color to the dot
age =dfp[:50]['age']

plt.xticks(rotation=90)
plt.scatter(x,y, s=100, c=age, cmap = 'Greens', edgecolor='black', alpha = .75)

# Code for clor bar
cbar = plt.colorbar()
cbar.set_label('Age')

#show the plot
plt.show()

```

```python
# Modifying the szie of the dots based on a value

# get the duration of the listing
time_listed = dfp[:50]['listing_duration']

age =dfp[:50]['age']
plt.xticks(rotation=90)

# use the s argument to have the size of the plot
plt.scatter(x,y, s=time_listed, c=age, cmap = 'Greens', edgecolor='black', alpha = .75)

cbar = plt.colorbar()
cbar.set_label('Age')

plt.show()
```

```python
# Using a log scale
plt.scatter(x,y, s=time_listed, c=age, cmap = 'summer', edgecolor='black', alpha = .75)
cbar = plt.colorbar()
cbar.set_label('Age')

# using log scales
plt.yscale('log')
# plt.xscale('log')

plt.show()
```
