```python
def extract_attr(x):
    try:
        return eval(x)[0]['value']
    except:
        return np.nan
    
```

```python
# df1['attrs'].value_counts()
```

```python
df1['Condition'] = df1['attrs'].apply(lambda x:extract_attr(x))
```

```python
def extract_price(x):
    return (eval(x))['value']    
```

```python
df1['Price']= df1['price_obj'].apply(lambda x: extract_price(x))
```
