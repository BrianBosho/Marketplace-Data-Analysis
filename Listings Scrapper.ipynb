```python
def get_listingrecursive(link):
    req = requests.get(link)
    adverts_list = []
    data = req.content
    data_dict = json.loads(data.decode('utf-8'))
    adverts = data_dict['adverts_list']['adverts']
    ads_list.extend(adverts)
    next_url = data_dict['next_url']
    if len(next_url)>10:
        get_listingrecursive(next_url)    
   
    return adverts_list
```
